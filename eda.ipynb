# Exploratory Data Analysis of Audible Books

## Introduction
Audible is the world's most widely used audiobook service. Provided by Amazon, this service provides over 200,000 audiobooks in a variety of subjects and languages. The Audible dataset contains information about over 87,000 audiobooks - including their title, author, time (in minutes), price, and ratings. The purpose of this exploratory data analysis is to gain insights into the audiobooks listed and inform our understanding of them.

## Methods
I have conducted a systematic and thorough exploration of the data, including data cleaning and preparation, descriptive statistics and visualization, and hypothesis testing. I used Python and its data analysis libraries, including Pandas, NumPy, Matplotlib, and Seaborn to analyze the data. Data cleaning was performed in Excel.


```python
#import libraries needed
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#import the audible data
data = pd.read_csv("audible_cleaned.csv")

#describe the main statistical measurements of each column of the data
data.describe()
```

```python
#describe the correlation between columns
data.corr()
```

```python
#describe the shape of the DataFrame
data.shape
```

```python
#heat map of the correlations
sns.heatmap(data.corr(), annot=True, cmap='magma')
plt.show()
```

```python
#distribution plot for the price of the audiobooks
sns.displot(data['price(USD)'], kde=True)
```

```python
#distribution plot of stars
sns.displot(data['stars'])
```

```python
#audiobook(s) with the maximum price
data[data['price(USD)'] == data['price(USD)'].max()]
```

```python
#audiobook(s) with the minimum price
data[data['price(USD)'] == data['price(USD)'].min()]
```

```python
#Maximum price of audiobooks across languages
data.groupby(['language']).max().sort_values(by='price(USD)', ascending=False)['price(USD)']
```

```python
#audiobooks with the maximum rating
data[data['stars'] == data['stars'].max()]
len(data[data['stars'] == data['stars'].max()])
```

```python
#mean rating for audiobooks
round(data['stars'].mean(), 2)
```

```python
#audiobooks with maximum price across ratings
data.groupby(['stars']).max().sort_values(by='stars', ascending=False)['price(USD)']
```

```python
#mean audiobook price across ratings
data.groupby(['stars']).mean().sort_values(by='stars', ascending=False)['price(USD)']
```

```python
#maximum rating of audiobooks across languages
data.groupby(['language']).max().sort_values(by='stars', ascending=False)['stars']
```

```python
#mean rating across languages
data.groupby(['language']).mean().sort_values(by='stars', ascending=False)['stars']
```
